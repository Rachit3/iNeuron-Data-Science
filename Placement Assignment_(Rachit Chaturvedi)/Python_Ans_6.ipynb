#Reading Pokemon Data(Same as question 3)

This first code is same as question 3, scroll down to look for insights asked

```python
import requests
import pandas as pd

url = 'https://raw.githubusercontent.com/Biuni/PokemonGO-Pokedex/master/pokedex.json'
response = requests.get(url)
json_data = response.json()

# Extract relevant data from JSON
pokemon_list = json_data['pokemon']
pokemon_data = []

for pokemon in pokemon_list:
    pokemon_data.append({
        'id': pokemon['id'],
        'num': pokemon['num'],
        'name': pokemon['name'],
        'img': pokemon['img'],
        'type': ', '.join(pokemon['type']),
        'height': pokemon['height'],
        'weight': pokemon['weight'],
        'candy': pokemon['candy'],
        'candy_count': pokemon.get('candy_count', None),
        'egg': pokemon['egg'],
        'spawn_chance': pokemon['spawn_chance'],
        'avg_spawns': pokemon['avg_spawns'],
        'spawn_time': pokemon['spawn_time'],
        'multipliers': ', '.join([str(multiplier) for multiplier in pokemon['multipliers']] if pokemon['multipliers'] else []),
        'weaknesses': ', '.join(pokemon['weaknesses']),
        'next_evolution': pokemon.get('next_evolution', None),  # Handle missing field
        'prev_evolution': pokemon.get('prev_evolution', None)
    })

# Convert extracted data to DataFrame
df = pd.DataFrame(pokemon_data)
df
```

**Pokemons whose spawn time is less than 300 seconds.** 👇 👇

```python
df['spawn_time']
```

```python
#Pokemons whose spawn time is less than 300 seconds.

def convert_spawn_time_to_seconds(spawn_time):
    if spawn_time == 'N/A' or ':' not in spawn_time: #since there are N/A values
        return None
    minutes, seconds = spawn_time.split(':')
    if not minutes.isdigit() or not seconds.isdigit():
        return None
    return int(minutes) * 60 + int(seconds)


df['spawn_time_seconds'] = df['spawn_time'].apply(convert_spawn_time_to_seconds)

filtered_df = df[df['spawn_time_seconds'] < 300][['id','name','type','spawn_time','spawn_time_seconds']]

filtered_df
```
